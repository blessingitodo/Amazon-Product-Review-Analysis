```python
import pandas as pd
import numpy as np
from datetime import datetime
import os
import json

folder_path = '/content/drive/MyDrive/Amazon Product Reviews'
files = os.listdir(folder_path)
jsonl_files = [os.path.join(folder_path, file)for file in files if file.endswith('.jsonl')]

jsonl_files
```

```python
# importing files into the dataframe

reviews_dfs = []
items_dfs = []
for file in jsonl_files:
  if 'meta_' in os.path.basename(file):
    items_dfs.append(pd.read_json(file, lines=True))
  else:
    reviews_dfs.append(pd.read_json(file, lines=True))

reviews = pd.concat(reviews_dfs, ignore_index=True) if reviews_dfs else pd.DataFrame()
items = pd.concat(items_dfs, ignore_index=True) if items_dfs else pd.DataFrame()

reviews.head()
```

```python
# function to strip all the columns in the df
def strip(df):
  for column in df.columns:
    if df[column].dtype == 'object':
      df[column] = df[column].str.strip()
strip(reviews)
```

```python
# data cleaning to drop irrelevant columns and rename some columns for readability
items = items.drop(['bought_together', 'subtitle', 'author'], axis=1)
items = items.rename({'parent_asin': 'parent_id'}, axis=1)
reviews = reviews.rename({'asin': 'product_id', 'parent_asin': 'parent_id'}, axis=1)

items.head()
```

```python
reviews.head()
```

```python
# percentage of each rating
percentage_rating = reviews.groupby('rating').agg({'rating': 'count'})
percentage_rating.columns = ['total']
percentage_rating = percentage_rating.reset_index()

percentage_rating['percentage'] = (percentage_rating['total'] / percentage_rating['total'].sum()) * 100

percentage_rating
```

```python
# Most and least reviewed product

min_max = reviews.groupby(['product_id']).agg({'user_id': ['count']})
min_max.columns = ['total']
min_max = min_max.reset_index()

min_max.sort_values(by='total', ascending=False)
```

```python
# average rating for each product

avg_rating = reviews.groupby(['product_id']).agg({'rating': ['mean']})
avg_rating.columns = ['average_rating']
avg_rating = avg_rating.reset_index()

avg_rating.sort_values(by='average_rating', ascending=False)
```

```python
# merging both tables in the dataset
merged = pd.merge(left=reviews, right=items, left_on=['parent_id'], right_on=['parent_id'], how='inner')
merged.head()
```

```python
# dropping irrelevant columns from merged
merged = merged.drop(['images_x', 'images_y', 'videos', 'categories'], axis=1)
merged.head()
```

```python
# how many reviews were verified purchases

verified_purchases = reviews.groupby(['verified_purchase']).agg({'user_id': 'count'})
verified_purchases.columns = ['total_purchased']
verified_purchases = verified_purchases.reset_index()

verified_purchases
```

```python
# How has the volume of reviews changed over time?, by year, months

reviews['timestamp'] = pd.to_datetime(reviews['timestamp'])
temporal_change = reviews.groupby([reviews['timestamp'].dt.year, reviews['timestamp'].dt.month]).agg({'user_id': 'count', 'rating': 'mean'})
temporal_change.columns = ['total_reviews', 'average_rating']
# temporal_change = temporal_change.reset_index()

temporal_change
```

```python
# changes in the number of reviews over the years and months
temporal_change = pd.pivot_table(reviews, index=reviews['timestamp'].dt.year, columns=reviews['timestamp'].dt.month,
                                 values=['user_id'], aggfunc='count')
temporal_change = temporal_change.reset_index()

temporal_change
```

```python
# changes in average rating over the years and months
temporal_change_rating = pd.pivot_table(reviews, index=reviews['timestamp'].dt.year, columns=reviews['timestamp'].dt.month,
                                 values=['rating'], aggfunc='mean')
temporal_change_rating = temporal_change_rating.reset_index()

temporal_change_rating
```

```python
# What is the average rating per product category

rating_per_category = merged.groupby(['main_category']).agg({'rating': 'mean'})
rating_per_category.columns = ['average_rating']
rating_per_category = rating_per_category.reset_index()

rating_per_category
```

```python
# Are there products with consistently high (or low) ratings

high_low_rating = merged.groupby(['parent_id']).agg({'rating': ['mean', 'max', 'min']})
high_low_rating.columns = ['average_rating', 'max_rating', 'min_rating']
high_low_rating = high_low_rating.reset_index()
high_low_rating.sort_values(by='average_rating', ascending=False)
```

```python
# What percentage of reviews include the maximum rating

percentage_max_rating = reviews.groupby('rating').agg({'rating': 'count'})
percentage_max_rating.columns = ['total']
percentage_max_rating = percentage_max_rating.reset_index()

percentage_max_rating['percentage'] = (percentage_max_rating['total'] / percentage_max_rating['total'].sum()) * 100

percentage_max_rating.sort_values(by='rating', ascending=False)
```

```python
# What is the distribution of ratings by day of the week?

day_of_week = reviews.groupby(reviews['timestamp'].dt.day_name()).agg({'rating': 'mean'})
day_of_week.columns = ['average_rating']
day_of_week = day_of_week.reset_index()

day_of_week.sort_values(by='average_rating', ascending=False)
```

```python
# What are the most common words in the review titles or text?
import re

all_title = " ".join(reviews['title'].dropna())
all_text = " ".join(reviews['text'].dropna())

cleaned_title = re.sub(r'[^\w\s]', '', all_title.lower())
cleaned_text = re.sub(r'[^\w\s]', '', all_text.lower())

title_words = cleaned_title.split()
text_words = cleaned_text.split()

title_word_count = pd.Series(title_words).value_counts().head(10)
text_word_count = pd.Series(text_words).value_counts().head(10)

print("Most Common Words in Title:")
print(title_word_count)

print("\nMost Common Words in Text:")
print(text_word_count)
```

```python
# Which products have improved or declined in average rating over time in 2023?

reviews_2023 = reviews[reviews['timestamp'].dt.year == 2023]
avg_rating_over_time = pd.pivot_table(reviews_2023, index='parent_id', columns=reviews_2023['timestamp'].dt.month, values='rating', aggfunc='mean')
avg_rating_over_time = avg_rating_over_time.reset_index()

avg_rating_over_time.sort_values(by=9, ascending=False)
```

```python
# How does the number of helpful votes correlate with the review rating?

helpful_rating_corr = reviews[['helpful_vote', 'rating']].corr()
helpful_rating_corr
```

```python
# What are the most common words or phrases in review titles or text for highly-rated and poorly-rated products? using SAC

high_rating = reviews[reviews['rating'] >= 4]
low_rating = reviews[reviews['rating'] < 4]

high_title = " ".join(high_rating['title'].dropna())
low_title = " ".join(low_rating['title'].dropna())


cleaned_high_title = re.sub(r'[^\w\s]', '', high_title.lower())

cleaned_low_title = re.sub(r'[^\w\s]', '', low_title.lower())

high_title_words = cleaned_high_title.split()
low_title_words = cleaned_low_title.split()

low_title_word_count = pd.Series(low_title_words).value_counts().head(10)

high_title_word_count = pd.Series(high_title_words).value_counts().head(10)

print("Most Common Words in High Rating Title:")
print(high_title_word_count)

print("\nMost Common Words in Low Rating Title:")
print(low_title_word_count)
```

```python
# Do reviews from verified purchases tend to have more helpful votes?

verified_helpful_votes = reviews.groupby('verified_purchase').agg({'helpful_vote': ['count']})
verified_helpful_votes.columns = ['total_helpful_votes']
verified_helpful_votes = verified_helpful_votes.reset_index()

verified_helpful_votes
```

```python
# Which reviews have the most helpful votes, and what patterns can be identified in these reviews (e.g., keywords, structure)

top_reviews = reviews.nlargest(20, 'helpful_vote')
top_reviews
```

```python
# How do users who review multiple products rate across different categories?

user_category_rating = merged.groupby(['user_id', 'main_category']).agg({'rating': 'mean'})
user_category_rating.columns = ['average_rating']
user_category_rating = user_category_rating.reset_index()

user_category_rating = user_category_rating.sort_values(by='average_rating', ascending=False)
user_category_rating.head(10)
```
